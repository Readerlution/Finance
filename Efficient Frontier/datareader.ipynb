```python
# import packages for research
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas_datareader as pdr
from datetime import datetime
%matplotlib inline
```

```python
start = datetime(2020,1,1)
end = datetime(2021,12,31)

종목 = '삼정전자'
```

```python
def get_code(df, name):
    code = df.query(f"name=='{name}'")['code'].to_string(index=False)
    code = code.strip()
    return code

code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download', header=0)[0]
code_df.rename(columns={'회사명': 'name', '종목코드':'code'}, inplace=True)
code_df.code = code_df.code.map('{:06d}'.format)
code = get_code(code_df, '삼성전자')
code = code + '.KS'
```

```python
df = pdr.DataReader(code, 'yahoo', start, end)
```
