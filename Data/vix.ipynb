```python
from pandas_datareader import data as pdr
import pandas_datareader as web
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import yfinance as yf
yf.pdr_override()
```

```python
vix = web.DataReader("VIXCLS", 'fred', start= "1970-01-01")
gold_vix = web.DataReader("GVZCLS", 'fred', start= "1970-01-01")
oil_vix = web.DataReader("OVXCLS", 'fred', start= "1970-01-01")
rus_vix = web.DataReader("RVXCLS", 'fred', start= "1970-01-01")
nq_vix = web.DataReader("VXNCLS", 'fred', start= "1970-01-01")
```

```python
vix_df = pd.concat([vix,gold_vix,oil_vix,rus_vix,nq_vix], axis=1)
```

```python
vix_df
```

```python
decile = np.linspace(0.1, 1.0, num=10)
vix_df.quantile(decile)
```

```python
# vix df without extreme values
vix_df_average = vix_df[vix_df < vix_df.quantile(decile).loc[0.9]]
vix_df_average.mean()
```

```python
vix_df_average.quantile(decile)
```
