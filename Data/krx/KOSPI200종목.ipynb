```python
from pykrx import stock
import pandas as pd
import numpy as np
```

```python
path = "D:\gdrive\코멘토\Kodex_200_20230721.xls"

df = pd.read_excel(path)
```

```python
df.columns = df.loc[1]
df = df.iloc[3:]
df.index = np.arange(1,len(df)+1)
df = df.drop(columns=["번호", "ISIN", "등락(원)"])
price_df = stock.get_market_ohlcv("20230721", market = "KOSPI")
close = price_df.loc[df["종목코드"]]["종가"]
```

```python
df = pd.merge(df, close, left_on="종목코드", right_on="티커", how="left")
```
